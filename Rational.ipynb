```
class Rational:
  def __init__(self, z, n = 1):
    self.z = z
    self.n = n
    self.normalize()
    #self.mixed()

  def __str__(self):
    return str(self.z) + "/" + str(self.n)
  
  def normalize(self):
    if self.n < 0:
      self.n *= -1
      self.z *= -1

    z, n = abs(self.z), self.n
    while (z % n != 0):
      z, n = n, z % n
    self.z //= n
    self.n //= n

  def __add__(self, other):
    chis = self.z * other.n + other.z * self.n
    znam = other.n * self.n
    b = Rational(chis, znam)
    return b

  def __iadd__(self, other):
    chis = self.z * other.n + other.z * self.n
    znam = other.n * self.n
    b = Rational(chis, znam)
    return b

  def __sub__(self, other):
    chis = self.z * other.n - other.z * self.n
    znam = other.n * self.n
    b = Rational(chis, znam)
    return b

  def __isub__(self, other):
    chis = self.z * other.n - other.z * self.n
    znam = other.n * self.n
    b = Rational(chis, znam)
    return b

  def __mul__(self, other):
    chis = self.z * other.z
    znam = self.n * other.n
    b = Rational(chis, znam)
    return b

  def __imul__(self, other):
    chis = self.z * other.z
    znam = self.n * other.n
    b = Rational(chis, znam)
    return b

  def __truediv__(self, other):
    chis = self.z * other.n
    znam = self.n * other.z
    b = Rational(chis, znam)
    return b

  def __idiv__(self, other):
    chis = self.z * other.n
    znam = self.n * other.z
    b = Rational(chis, znam)
    return b

  def __pow__(self, other):
    chis = self.z ** other
    znam = self.n ** other
    b = Rational(chis, znam)
    return b
  
  def __neg__(self):
    chis = -self.z
    znam = -self.n
    b = Rational(chis, znam)
    return b
  def __abs__(self):
    b = Rational(abs(self.z), abs(self.n))
    return b

  def __lt__(self, other):
    return ((self.z / self.n) < (other.z / other.n))

  def __gt__(self, other):
    return ((self.z / self.n) > (other.z / other.n))

  def __le__(self, other):
    return ((self.z / self.n) <= (other.z / other.n))

  def __ge__(self, other):
    return ((self.z / self.n) >= (other.z / other.n))

  def __eq__(self, other):
    return ((self.z / self.n) == (other.z / other.n))

  def __ne__(self, other):
    return ((self.z / self.n) != (other.z / other.n))

  def mixed(self):
    if(self.z > self.n):
      full = self.z // self.n
      chis = self.z - self.n
      znam = self.n
    b = Rational(chis, znam)
    return str(full) + " Ð¸ " + str(b)
```
